### Histogram


*   **`bins`**: Bins are the intervals into which the data is divided for counting and display in the histogram.
    - Having 20 bins means the data range will be split into 20 intervals
*   **`alpha`**: controls the transparency of the bars.
    - A value of 1.0 would be completely opaque, while 0.0 would be completely transparent.



```python
# Histogram - plot 1
rng = np.random.RandomState(12345) #setting seed number, so everyone will get the same number
plt.hist(rng.standard_normal(100), bins=20, color='black', alpha=0.8);
#plt - means to plot
#.hist - means calling out the function
#
#bins=20 : chop the number into 20 bins, from all the number that we have
#color='black'
#alpha =0.8 - means opacity of the color
```

### **linestyle**
1. some other options for the linestyle argument:

*   `'-'` (**solid**): This is the default linestyle and creates a solid line.
*   `'--'` (**dashed**): Creates a dashed line.
*   `'-.'` (**dash-dot**): Creates a line with alternating dashes and dots.
*   `':'` (**dotted**): Creates a dotted line.
*   `'**None**'` or `' '` or `''` (no line): This will plot only markers if specified, or no line at all.

2. we can also use the shorter codes for these linestyles:
*   **`solid`**: `-`
*   **`dashed`**: `--`
*   **`dashdot`**: `-.`
*   **`dotted`**: `:`

```python
# Line plot : plot 3
rng = np.random.RandomState(12345)
plt.plot(rng.standard_normal(50).cumsum(), color='black', linestyle='dashed');
#
```

### **Place all the subplots into one panel**

1.   **plt.figure()** - create a new figure
2.   **add_subplot** - to create one or more subplots
3.   **Subplots** are numbered sequentially **starting from 1, going from left to right and top to bottom.**
    - a **(2, 2, 1)** argument means we're adding the first subplot (top-left corner) in the 2x2 grid.


```python
# Create one or more subplots
# 2 x 2; 4 plots in total
# Plot axis
## check  page 372 - figures and subplots
rng = np.random.RandomState(12345) #set seed number
fig = plt.figure(dpi=300) #plot the figure, dpi is the resolution, usually is 300
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(rng.standard_normal(100), bins=20, color='black', alpha=0.3); #coding for ax1
ax2.scatter(np.arange(30), np.arange(30) + 3 * rng.standard_normal(30)); #coding for ax2
ax3.plot(rng.standard_normal(50).cumsum(), color='black', linestyle='dashed'); #coding for ax3

# To save the figures
plt.savefig('/content/drive/MyDrive/STQD6014/combi3.pdf')
#/content/drive/MyDrive/STQD6014 - this is the path name
#need to put single quote, then put /filename. in example above, filename is combi3
```

### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# Change the value of wspace and hspace
rng = np.random.RandomState(12345)
fig, axes = plt.subplots(2, 2, sharex=True, sharey=False)
#what is sharex & share y?

# Using a nested loop to populate 2x2 grid
for i in range(2): #
    for j in range(2):
        axes[i, j].hist(rng.standard_normal(500), bins=50,
                        color='black', alpha=0.5)

# Adjusting the spacing
fig.subplots_adjust(wspace=0.3, hspace=0.3)
#spacing between the subplots, wspace = wide spacing, hspace = height spcing
```

### ***Setting the title, axis labels, ticks, and tick labels***

```python
# An example
rng = np.random.RandomState(12345)
fig, ax = plt.subplots()
ax.plot(rng.standard_normal(1000).cumsum())

# Changex x-axis ticks annotation
ax.set_xticks([0, 250, 500, 750,1000])

# Cabel the x-axis ticks with names
ax.set_xticklabels(['one','two', 'three', 'four','five'], rotation=30, fontsize=18);
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
ax.set_title('My first matplotlib plot');
```

### **Annotations and Drawing on a Subplot**

```python
# The SPX data
from datetime import datetime

fig, ax = plt.subplots()

data = pd.read_csv('https://bit.ly/3iofIY8', index_col=0, parse_dates=True)
#data
spx = data['SPX']
#spx
spx.plot(ax=ax, color='blue')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')]

for date, label in crisis_data:
  ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis');
```

### **Adding shapes on the same plot**

```python
#THIS WILL COME UP IN EXAM
# Adding shapes to the same plot
fig, ax = plt.subplots()

#create the shape
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
#rect = plt.Rectangle((#the starting point x,y), the width, height, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
#circ = plt.Circle((radius xy), middle point, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.5]], color='green', alpha=0.5)

#add shape to the plot using ax.addpatch(shp)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
```

```python
# Challenge: plot a pentagon, ellipse, square figures in the exact same figure
fig, ax = plt.subplots()

# Pentagon
  # Center, sides, radius
pentagon = patches.RegularPolygon((0.2, 0.7), radius=0.1, color='red')
ax.add_patch(pentagon)

# Ellipse
  # Center, width, height
ellipse = patches.Ellips((0.2, 0.7), 0.2, 0.1, color='navy')
ax.add_patch(ellipse)

# Square
  # Lower-left, width, height
square = patches.Rectangle((0.4, 0.2), 0.2, 0.2, color='green')
ax.add_patch(square)

plt.show()  # Display the figure
```

```python
# Adding shapes to a plot
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='black', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='blue', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='green', alpha=0.5)
sqr = plt.Rectangle((0.3, 0.1), 0.2, 0.2, color='black', alpha=0.3)

# Add ellipse
ellipse = mpatches.Ellipse((0.5,0.5), 0.2, 0.1)

# Add regular hex
hex = mpatches.RegularPolygon((0.7,0.8), 6, radius=0.1)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```
